```python
%pylab inline
plt.style.use("bmh")
import numpy as np
import pandas as pd
```

```python
stations = pd.read_csv('station.csv')
trip = pd.read_csv('trip.csv',parse_dates=["start_date"])
```

```python
# keep only trip where start_station_id != end_station_id
trip = trip[trip.start_station_id!=trip.end_station_id]
```

```python
# keep only id,name,lat,long
stations = stations[['id','name','lat','long']]
```

```python
# join trip data with stations
data = pd.merge(trip,stations,left_on='start_station_id',right_on='id')
data = pd.merge(data,stations,left_on='end_station_id',right_on='id',suffixes=('_from','_to'))
```

```python
# count number of trips between stations and rename columns
trips_freqs = data.groupby(['start_station_name','end_station_name']).size().reset_index().sort_values(by=0,ascending=False)
trips_freqs.rename(columns={'start_station_name':'source','end_station_name':'target',0: 'freq'}, inplace=True)
```

```python
# keep only routes with frequency >= 500
trips_freqs = trips_freqs[trips_freqs.freq >=500]
```

```python
# install networkx package
#!pip install networkx
import networkx as nx
```

```python
g = nx.Graph()
```

```python
for i, elrow in trips_freqs.iterrows():
    g.add_edge(elrow[0], elrow[1], attr_dict=elrow[2:].to_dict())
```

```python
plt.figure(figsize=(20, 16))
nx.draw(g, 
        pos=nx.fruchterman_reingold_layout(g,k=3/math.sqrt(g.order())),
        edge_color='black',
        node_color='orange', 
        node_size=200, 
        with_labels=True,
        alpha=0.75,
        font_size=15,
        width= np.log(trips_freqs.freq)/10)
plt.title('Graph Representation of Sleeping Giant Trail Map', size=15)
plt.show()
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
graph = nx.from_pandas_edgelist(trips_freqs, 
                                 source = 'source',
                                 target = 'target',
                                 edge_attr = 'freq',
                                 create_using = nx.DiGraph())
```

```python
#!pip install networkx
import networkx as nx

# count number of trips between stations and rename columns
trips_freqs = data.groupby(['start_station_name','end_station_name']).size().reset_index().sort_values(by=0,ascending=False)
trips_freqs.rename(columns={'start_station_name':'source','end_station_name':'target',0: 'freq'}, inplace=True)

g = nx.Graph()
for i, elrow in trips_freqs.iterrows():
    g.add_edge(elrow[0], elrow[1], attr_dict=elrow[2:].to_dict())

fig=plt.figure(figsize=(20, 16))
nx.draw(g, 
        pos=nx.fruchterman_reingold_layout(g,k=3/math.sqrt(g.order())),
        edge_color='white',
        node_color='orange', 
        node_size=200, 
        alpha=0.75,
        font_size=15,
        width= np.log(trips_freqs.freq+1)/10
       )
fig.set_facecolor("#222222")
plt.title('Graph Representation of Bikes Routes', size=15, color='white')
plt.show()    
```

```python
import plotly.offline as ply
scl = [ [0,"rgb(5, 10, 172)"],[0.35,"rgb(40, 60, 190)"],[0.5,"rgb(70, 100, 245)"],\
    [0.6,"rgb(90, 120, 245)"],[0.7,"rgb(106, 137, 247)"],[1,"rgb(220, 220, 220)"] ]

merged = pd.merge(stations[['name', 'id', 'lat', 'long']], trip[['start_station_id']], how='left', left_on=['id'],
                  right_on=['start_station_id'])

# x2 = merged.groupby(['start_station_id']).size().reset_index(name='counts')

x2 = merged.groupby(['name', 'lat', 'long']).size().reset_index(name='counts')
x2['text'] = x2.apply(lambda x: 'name:{0} countt:{1}'.format(x2['name'], x2['counts']))

x2['text'] = 'Name:'+ x2['name'] + '\nSize:' + x2['counts'].astype(str)



data = [ dict(
        type = 'scattergeo',
        locationmode = 'country names',
        lon = x2['long'],
        lat = x2['lat'],
        text = x2['text'],
        mode = 'markers',
        marker = dict(
            size = 8,
            opacity = 0.8,
            reversescale = True,
            autocolorscale = False,
            symbol = 'square',
            line = dict(
                width=1,
                color='rgba(102, 102, 102)'
            ),
            colorscale = scl,
            cmin = 0,
            color = x2['counts'],
            cmax = x2['counts'].max(),
            colorbar=dict(
                title="Stations demand"
            )
        ))]

layout = dict(
        title = 'Stations demand',
        #colorbar = True,
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showland = True,
            landcolor = "rgb(250, 250, 250)",
            subunitcolor = "rgb(217, 217, 217)",
            countrycolor = "rgb(217, 217, 217)",
            countrywidth = 0.5,
            subunitwidth = 0.5
        ),
    )


fig = dict( data=data, layout=layout )

ply.offline.plot( fig)
```

```python
import colorlover as cl
import plotly.graph_objs as go

chosen_colors=cl.scales['5']['qual'][np.random.choice(list(cl.scales['5']['qual'].keys()))]

trip_count_by_date=trip.groupby(['date'])['id'].count().reset_index()



trace1 = go.Scatter(
    x=trip_count_by_date.date,
    y=trip_count_by_date.id,
    mode='markers',
    line=dict(
        color=chosen_colors[0]
    ),
    name='Daily'
)

data=[trace1]

layout = go.Layout(
    title='Number oftrips per day',
    xaxis=dict(
        title='Date',
        showgrid=False
    ),
    yaxis=dict(
        title='Number oftrips'
    ),
    hovermode='closest',
)

figure = go.Figure(data=data, layout=layout)

ply.plot(figure)
```
